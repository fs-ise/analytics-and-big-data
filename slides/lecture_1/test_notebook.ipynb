# A New Headquarter Location in Frankfurt

**Business question**

Where should we locate a new headquarters in Frankfurt?

Assumption: employees value **good food nearby**.

We compare locations based on:
- office space
- organic restaurants nearby

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("frankfurt_locations.csv")

plt.scatter(df.office_size, df.restaurants)

for _, r in df.iterrows():
    plt.annotate(r.location, (r.office_size, r.restaurants))

plt.xlabel("Office space (sqm)")
plt.ylabel("Nearby restaurants")
plt.title("Candidate HQ locations in Frankfurt")

plt.show()
```

## Recommendation

Optimal location: **Westend**  (best mix of office space and restaurants)

Fallback option: **Ostend**.
